# PhishScan – Email Phishing Detection using Machine Learning

This notebook documents the development of the phishing detection model used in the PhishScan web application. It covers data preprocessing, feature engineering, model training, evaluation, and serialization of the final model and vectorizer.


## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

## 2. Load Dataset

```python
df = pd.read_csv("../datasets/phishing.csv")

# Show columns to verify
print("Columns:", df.columns)

# Rename and keep only what's needed
df = df[['Email Text', 'Email Type']]
df.columns = ['email_text', 'label']

# Map all non-phishing emails as 0
df['label'] = df['label'].apply(lambda x: 1 if 'phish' in x.lower() else 0)

# Drop missing values
df.dropna(inplace=True)

print("Label counts:\n", df['label'].value_counts())
df.head()
```

## 3. Text Preprocessing

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_email(text):
    if not isinstance(text, str) or pd.isna(text):
        return ""
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    words = word_tokenize(text.lower())
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words and len(w) > 2]
    return " ".join(words)

df['cleaned_text'] = df['email_text'].apply(clean_email)
df = df[df['cleaned_text'].str.strip().astype(bool)]
```

## 4. Feature Engineering

```python
extra_data = pd.DataFrame({
    "email_text": [
        "Dear user, we noticed suspicious login attempts. Please reset your password: http://secure-login.com",
        "Hello, your invoice is ready. Confirm payment here: http://quick-pay-invoice.com",
        "You won a free iPhone! Click below to claim your prize: http://freeiphone.fakeurl",
        "Your PayPal account has been locked. Login to verify: http://paypal-security-check.com"
    ],
    "label": [1, 1, 1, 1]
})

extra_data['cleaned_text'] = extra_data['email_text'].apply(clean_email)
df = pd.concat([df, extra_data], ignore_index=True)
df = df[df['cleaned_text'].str.strip().astype(bool)]
```

```python
phishing_keywords = [
    "reset your password", "verify your account", "click below", "login to confirm",
    "update billing", "urgent", "account suspended", "invoice", "free", "claim", "congratulations",
    "limited time", "security alert", "confirm your identity", "pay now", "alert", "secure link"
]

def keyword_flag(text):
    text = text.lower()
    return int(any(kw in text for kw in phishing_keywords))

df['keyword_flag'] = df['email_text'].apply(keyword_flag)
```

```python
vectorizer = TfidfVectorizer(max_features=3000)
X_text = vectorizer.fit_transform(df['cleaned_text']).toarray()
X = np.hstack((X_text, df['keyword_flag'].values.reshape(-1, 1)))
y = df['label'].values
```

## 5. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## 6. Model Training

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

## 7. Evaluation

```python
y_pred = model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
importances = model.feature_importances_
top_indices = np.argsort(importances)[-20:]

plt.figure(figsize=(10, 6))
plt.barh(range(len(top_indices)), importances[top_indices])
plt.yticks(range(len(top_indices)), [vectorizer.get_feature_names_out()[i] for i in top_indices])
plt.xlabel("Importance Score")
plt.title("Top 20 Important Words")
plt.tight_layout()
plt.show()
```

```python
import requests
import re

VT_API_KEY = "b3bcf57674026967cc3e583bfab96a559ca9af16ec26edf69ed5005d35cd23be"

def extract_links(text):
    return re.findall(r'http[s]?://\S+', text)

def check_virustotal(url):
    headers = {"x-apikey": VT_API_KEY}
    response = requests.get(f"https://www.virustotal.com/api/v3/urls", params={"url": url}, headers=headers)
    if response.status_code == 200:
        data = response.json()
        analysis_id = data["data"]["id"]
        report = requests.get(f"https://www.virustotal.com/api/v3/analyses/{analysis_id}", headers=headers)
        if report.status_code == 200:
            stats = report.json()["data"]["attributes"]["stats"]
            return stats
    return None
```

```python
def predict_email(text):
    cleaned = clean_email(text)
    vector = vectorizer.transform([cleaned]).toarray()
    keyword = keyword_flag(text)
    final_input = np.hstack((vector, np.array([[keyword]])))
    prediction = model.predict(final_input)

    print("Prediction:", "⚠️ Phishing" if prediction[0] == 1 else "✅ Legitimate")

    # --- VirusTotal Link Checking ---
    links = extract_links(text)
    if links:
        print("\nFound links:")
        for link in links:
            print("🔗", link)
            try:
                headers = {"x-apikey": VT_API_KEY}
                # Submit the URL to get analysis ID
                url_id = requests.post("https://www.virustotal.com/api/v3/urls", headers=headers, data={"url": link})
                if url_id.status_code == 200:
                    analysis_id = url_id.json()["data"]["id"]
                    # Get analysis report
                    report = requests.get(f"https://www.virustotal.com/api/v3/analyses/{analysis_id}", headers=headers)
                    if report.status_code == 200:
                        stats = report.json()["data"]["attributes"]["stats"]
                        print(f"🛡️ VirusTotal scan for {link}: {stats}")
                    else:
                        print("⚠️ Could not retrieve scan report.")
                else:
                    print("⚠️ Could not submit URL to VirusTotal.")
            except Exception as e:
                print("❌ Error checking VirusTotal:", str(e))
    else:
        print("\nNo links found in the email.")
```
