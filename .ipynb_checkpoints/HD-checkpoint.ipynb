```python
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # remove HTML
    tokens = nltk.word_tokenize(text.lower())
    return " ".join(lemmatizer.lemmatize(token) for token in tokens if token.isalpha())

# Apply text cleaning to the email body column (adjust column name if needed)
df['cleaned_body'] = df['Email Text'].apply(clean_text)
```

```python
print(df.columns.tolist())
```

```python
# TF-IDF Vectorization on cleaned email text
tfidf = TfidfVectorizer(max_features=3000)
X_content = tfidf.fit_transform(df['cleaned_body'])

# Target labels from 'Email Type' column (assuming 1 = phishing, 0 = legit)
y = df['Email Type']

# If no metadata like 'From', just use content features
X_combined = X_content
```

```python
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_combined, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, stratify=y_resampled, random_state=42
)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
feature_importances = model.feature_importances_
indices = np.argsort(feature_importances)[-20:]  # Top 20 important features

plt.figure(figsize=(10,6))
plt.barh(range(len(indices)), feature_importances[indices])
plt.yticks(range(len(indices)), [tfidf.get_feature_names_out()[i] for i in indices])
plt.xlabel('Importance')
plt.title('Top 20 Important Words')
plt.tight_layout()
plt.show()
```

```python
def predict_email(text):
    cleaned = clean_text(text)
    text_vector = tfidf.transform([cleaned])
    prediction = model.predict(text_vector)
    return "Phishing" if prediction[0] == 1 else "Legitimate"

# Example usage
test_email = "Your bank account has been compromised. Click here to verify your credentials immediately."
print("Prediction:", predict_email(test_email))
```

# 1. Classical ML: TF-IDF + Random Forest

```python
import pandas as pd
import re
import nltk
from nltk.stem import WordNetLemmatizer

# Download required NLTK data
nltk.download('punkt')
nltk.download('wordnet')

# Load full dataset and reduce to 1000 rows
df_full = pd.read_csv("datasets/Phishing_Email.csv")
df = df_full.sample(n=1000, random_state=42).reset_index(drop=True)  # Random 1000 emails

# Define text cleaning function
lemmatizer = WordNetLemmatizer()
def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # Remove HTML
    tokens = nltk.word_tokenize(text.lower())
    return " ".join(lemmatizer.lemmatize(token) for token in tokens if token.isalpha())

# Create 'cleaned_body' column from 'Email Text'
df["cleaned_body"] = df["Email Text"].apply(clean_text)

# Check
print(df[["Email Text", "cleaned_body", "Email Type"]].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd

X = TfidfVectorizer(max_features=3000).fit_transform(df["cleaned_body"])
y = df["Email Type"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# ✅ 2. Deep Learning: LSTM (Keras)

```python
# STEP 1: Import libraries
import pandas as pd
import numpy as np
import re
import nltk
import tensorflow as tf
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# STEP 2: Download NLTK resources
nltk.download('punkt')
nltk.download('wordnet')

# STEP 3: Load dataset and sample 1000 rows for speed
df = pd.read_csv("datasets/Phishing_Email.csv").sample(n=1000, random_state=42).reset_index(drop=True)

# STEP 4: Clean text using lemmatization
lemmatizer = WordNetLemmatizer()
def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # Remove HTML
    tokens = nltk.word_tokenize(text.lower())
    return " ".join(lemmatizer.lemmatize(token) for token in tokens if token.isalpha())

df["cleaned_body"] = df["Email Text"].apply(clean_text)

# STEP 5: Tokenize and pad sequences
texts = df["cleaned_body"]
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(texts)
X = pad_sequences(tokenizer.texts_to_sequences(texts), maxlen=200)

# STEP 6: Encode labels (phishing vs ham)
labels = LabelEncoder().fit_transform(df["Email Type"])
y = np.array(labels)

# STEP 7: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# STEP 8: Build LSTM model
model = Sequential([
    Embedding(input_dim=5000, output_dim=64),
    LSTM(64),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# STEP 9: Train the model
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

# 4. Anomaly Detection: Isolation Forest

```python
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

# NLTK setup
nltk.download('punkt')
nltk.download('wordnet')

# Load and clean data
df = pd.read_csv("datasets/Phishing_Email.csv")
lemmatizer = WordNetLemmatizer()
def clean_text(text):
    text = re.sub(r"<.*?>", "", str(text))
    tokens = nltk.word_tokenize(str(text).lower())
    return " ".join(lemmatizer.lemmatize(token) for token in tokens if token.isalpha())

df["cleaned_body"] = df["Email Text"].apply(clean_text)
df["label"] = LabelEncoder().fit_transform(df["Email Type"])  # 1 = phishing, 0 = ham

# TF-IDF features
vectorizer = TfidfVectorizer(max_features=3000)
X = vectorizer.fit_transform(df["cleaned_body"])

# Isolation Forest
model = IsolationForest(contamination=0.3, random_state=42)
model.fit(X)
raw_preds = model.predict(X)
predictions = [1 if x == -1 else 0 for x in raw_preds]  # 1 = phishing (anomaly), 0 = ham

# Report
print(classification_report(df["label"], predictions, target_names=["Ham", "Phishing"]))

# Confusion Matrix
conf_matrix = confusion_matrix(df["label"], predictions)
plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Phishing"], yticklabels=["Ham", "Phishing"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Isolation Forest")
plt.tight_layout()
plt.show()

# PCA for 2D projection
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X.toarray())

# Scatter plot with ground truth
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:,0], y=X_reduced[:,1], hue=df["label"], palette=["green", "red"], alpha=0.6)
plt.title("True Labels: Ham vs Phishing")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Actual", labels=["Ham", "Phishing"])
plt.tight_layout()
plt.show()

# Scatter plot with predicted labels
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:,0], y=X_reduced[:,1], hue=predictions, palette=["blue", "orange"], alpha=0.6)
plt.title("Isolation Forest Predicted Labels")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Predicted", labels=["Ham", "Phishing"])
plt.tight_layout()
plt.show()
```

# 5. Hybrid Model: TF-IDF + Metadata + Visualizations

```python
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import hstack

# Preprocess
nltk.download("punkt")
nltk.download("wordnet")
df = pd.read_csv("datasets/Phishing_Email.csv")

lemmatizer = WordNetLemmatizer()
def clean_text(text):
    text = re.sub(r"<.*?>", "", str(text))
    tokens = nltk.word_tokenize(text.lower())
    return " ".join(lemmatizer.lemmatize(token) for token in tokens if token.isalpha())

df["cleaned_body"] = df["Email Text"].apply(clean_text)
df["label"] = LabelEncoder().fit_transform(df["Email Type"])  # 1 = phishing

# TF-IDF + metadata
tfidf = TfidfVectorizer(max_features=3000)
X_text = tfidf.fit_transform(df["cleaned_body"])
df["length"] = df["cleaned_body"].apply(len)
X_combined = hstack([X_text, df[["length"]].values])
y = df["label"]

# Split & train
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, stratify=y)
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Report
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Phishing"], yticklabels=["Ham", "Phishing"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Random Forest (TF-IDF + Metadata)")
plt.tight_layout()
plt.show()

# PCA visual
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_combined.toarray())

# Actual
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:,0], y=X_reduced[:,1], hue=y, palette=["green", "red"], alpha=0.6)
plt.title("Actual Labels (0=Ham, 1=Phishing)")
plt.tight_layout()
plt.show()

# Predicted
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:,0], y=X_reduced[:,1], hue=model.predict(X_combined), palette=["blue", "orange"], alpha=0.6)
plt.title("Predicted Labels by Random Forest")
plt.tight_layout()
plt.show()
```
